```python
import joblib
import pandas as pd

# Load everything
model = joblib.load("../models/world_cup_predictor.pkl")
team_encoder = joblib.load("../models/team_encoder.pkl")
target_encoder = joblib.load("../models/target_encoder.pkl")

print("Everything loaded successfully!")
```

```python
print("Brazil" in team_encoder.classes_)
print("Argentina" in team_encoder.classes_)
```

```python
home_team = team_encoder.transform(["Brazil"])[0]
away_team = team_encoder.transform(["Argentina"])[0]

print(home_team)
print(away_team)
```

```python
sample = pd.DataFrame([{
    "home_team": home_team,
    "away_team": away_team,
    "neutral": True,
    "tournament_importance": 5,

    "home_matches": 100,
    "away_matches": 100,

    "home_win_rate": 0.70,
    "away_win_rate": 0.68,

    "home_avg_goals": 2.1,
    "away_avg_goals": 2.0,

    "home_avg_goals_conceded": 0.9,
    "away_avg_goals_conceded": 1.0,

    "home_recent_form": 2.4,
    "away_recent_form": 2.2,

    "home_goal_diff": 1.2,
    "away_goal_diff": 1.0
}])
```

```python
prediction = model.predict(sample)

print("Prediction:", target_encoder.inverse_transform(prediction)[0])
```

```python
probabilities = model.predict_proba(sample)[0]

for label, prob in zip(target_encoder.classes_, probabilities):
    print(f"{label}: {prob:.2%}")
```
